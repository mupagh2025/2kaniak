```python
import pint
import scipy
import matplotlib.pyplot as plt
import numpy as np
# !pip install open_atmos_jupyter_utils
from open_atmos_jupyter_utils import show_plot
```

```python
si = pint.UnitRegistry()

def calcDensity(p, T, R):
    return p / (R * T)
```

```python
R_air = scipy.constants.R * si.J / si.mole / si.K / (28.966 * si.kg / si.kmol)

airDens = calcDensity(1000 * si.hPa, 300 * si.K, R_air)

print(f'{airDens.to_base_units():.2f}')

print(f'{airDens.to(si.g/si.cm**3):.2e}')

print(f'{airDens.to(si.lbs / si.inch**3):.2e}')
```

```python
si.setup_matplotlib()

p = 1013.25 * si.hPa

plt.figure(dpi=200)

plt.plot(T:= np.linspace(200, 300) * si.K, rho:=calcDensity(p, T, R_air).to_base_units(), label='Density')

plt.legend()

# T = T.to(si.celsius)

plt.xlabel(f'Temperature [{T.units:~}]')
plt.ylabel(f'Density [{(rho.to_base_units()).units:~}]')

# plt.gca().xaxis.set_units(si.celsius)

# show_plot()
```
